```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
```

```python
head_ =["class",
        "Alcohol",
        "lic acid",
        "Ash",
        "Alcalinity of ash",
        "Magnesium",
        "Total phenols",
        "Flavanoids",
        "Nonflavanoid phenols",
        "Proanthocyanins",
        "Color intensity",
        "Hue",
        "diluted wines",
        "Proline"]
```

```python
data = pd.read_csv('wine.data', names = head_)
```

```python
data
```

```python
y = data["class"]
```

```python
X = data[["Alcohol",
        "lic acid",
        "Ash",
        "Alcalinity of ash",
        "Magnesium",
        "Total phenols",
        "Flavanoids",
        "Nonflavanoid phenols",
        "Proanthocyanins",
        "Color intensity",
        "Hue",
        "diluted wines",
        "Proline"]]
```

```python
kf = KFold(n_splits=5,shuffle=True,random_state=42)
kf.get_n_splits(X)
```

```python
X.loc[[1,4,3]]
```

```python
for train_index, test_index in kf.split(X):
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    for i in range(50):
        n_ne = i + 1
        neigh = KNeighborsClassifier(n_neighbors=n_ne)
        neigh.fit(X_train, y_train)
       # print(i)
        #acc_ = cross_val_score(neigh,X_test,y_test,scoring='accuracy')
        #print(acc_)
```

```python
acc = 0
saveK = 0
for i in range(50):
    n_ne = i + 1
    neigh = KNeighborsClassifier(n_neighbors=n_ne)
    acc_ = cross_val_score(neigh,X,y,cv=kf,scoring='accuracy')
    mean_quality = np.mean(acc_)
    if mean_quality > acc:
        acc = mean_quality
        saveK = n_ne
    print(n_ne," ",np.mean(acc_))
print(acc,saveK)
```
